How to rename column names

```python
df=df.rename(columns={'Mat':'Match', 
                      'Inns':'Innings',
                     'NO':'NotOut',
                     'HS':'Highest_score',
                     'Ave':'Average',
                     'BF':'Balls_Faced',
                     'SR':'Strike_Rate',
                     100:'Centuries',
                     50:'Half_Centuries',
                     0:'Ducks',
                     '4s':'Fours',
                     "6s":"Sixes"})
display(df.head(3))
```

How to split a column and create two new columns?

```python
df[["Player_Name","Country"]]=df['Player'].str.split("(",expand=True)
display(df.head(3))
display(df.shape)
```
