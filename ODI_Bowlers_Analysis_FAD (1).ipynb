8. How many players played for ICC?

```python
print(df.columns)
df[["Player_Name", "Country"]] = df['Player'].str.split("(", expand=True)
df = df.drop('Player', axis=1)
df['Country'] = df['Country'].str.replace(")", "")
new_col_sequence = ['Player_Name','Country','Span', 'Match', 'Innings', 'Balls', 'Runs', 'Wickets', 'Average','Economy', 'Strike_Rate', 'Four_Wickets', 'Five_Wickets']
df = df[new_col_sequence]
display(df.head(4))
```

```python
def icc_check(x):
    if "ICC" in x:
        return "Yes"
    else:
        return "No"
```

```python
def asia_check(x):
    if "Asia" in x:
        return "Yes"
    else:
        return "No"
```

```python
def africa_check(x):
    if "Afr" in x:
        return "Yes"
    else:
        return "No"
```

```python
df['played_for_ICC']=df['Country'].apply(icc_check)
df['played_for_Asia']=df['Country'].apply(asia_check)
df['played_for_Africa']=df['Country'].apply(africa_check)
display(df.head(3))
```

```python
print(df['played_for_ICC'].value_counts())
```

9. How many different countries are present in this dataset?

```python
df['Country'] = df['Country'].str.replace("/ICC","")
df['Country']=df['Country'].str.replace("ICC/","")
df['Country']=df['Country'].str.replace("Asia/","")
df['Country']=df['Country'].str.replace("Afr/","")
df['Country'].value_counts()
```

10. Which player(s) had played for the longest period of time?

```python
df['start_year']=df['Span'].str[0:4]
df['end_year']=df['Span'].str[5:]
df = df.drop("Span", axis=1)
df['start_year'] = df['start_year'].astype('int') 
df['end_year'] = df['end_year'].astype('int')
# print(df.info())
# display(df.head())
df['years_played'] = df['end_year'] - df['start_year']
df = df.drop(['start_year', "end_year"], axis=1)
# display(df.head(4))
df.sort_values(by='years_played', ascending = False).head(4)
```
